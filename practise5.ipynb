```python
y1=df[['C11','C22','C33','C12','C13','C23','C44','C55','C66']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X1,y1,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=500,max_depth=30)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred=rfc.predict(X_test)
```

```python
y_test
```

```python
z2=pd.DataFrame(rfc_pred)
```

```python
z2
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test,rfc_pred))
```

```python
print(metrics.mean_squared_error(y_test,rfc_pred))
```

```python
print(metrics.explained_variance_score(y_test,rfc_pred))
```
